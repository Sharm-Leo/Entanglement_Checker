# Entanglement Checker

This notebook provides a utility function to determine whether a given 
2-qubit quantum statevector is **entangled** or **separable**.

---

## 🔎 How it works
1. **Input**: A 4-dimensional statevector (from Qiskit or NumPy).
2. **Reshape**: The statevector is reorganized into a 2×2 matrix representation.
3. **SVD**: Singular Value Decomposition (SVD) is applied to this matrix.
4. **Interpretation**:  
   - If only **one singular value** is nonzero → the state is **separable**.  
   - If **more than one singular value** is nonzero → the state is **entangled**.
5. **Output**: A boolean value indicating entanglement.

---

## Note
- This method works for **pure states** (statevectors).

---

## 📘 Background
This method is based on the **Schmidt decomposition criterion** for bipartite pure states.  
It leverages the fact that entanglement in a 2-qubit system can be detected by examining the rank of the reshaped statevector.

---

## ✅ Usage
- Import the function into your workflow.
- Pass a 2-qubit statevector (from Qiskit or NumPy).
- The function will return `True` if entangled, `False` otherwise.


```python
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
# state_1 = Statevector([1/np.sqrt(2), 0, 0, -1/np.sqrt(2)])
state_1 = Statevector([1, 0, 0, 0])

np.shape(state_1)
```

```python
state_matrix = np.reshape(state_1,shape=[2,2])
state_matrix
```

```python
svd = np.linalg.svdvals(state_matrix)
svd
```

```python
rank = np.count_nonzero(svd > 0)

if rank > 1:
    print("2 Qubits are Entangled") 
else:
    print("2 Qubits are Seperable") 
```

# Understanding Rank and Entanglement

When working with a 2-qubit statevector, we reshape the 4-element vector 
into a 2×2 matrix. The rank of this matrix tells us whether the state 
is separable or entangled.

---

## 🔎 Why Rank Matters
- **Rank 1**:  
  The matrix can be written as an outer product of two vectors.  
  → This means the state can be expressed as a tensor product of two 
  single-qubit states.  
  → The qubits are **separable** (not entangled).

- **Rank 2**:  
  The matrix cannot be written as a single outer product.  
  → More than one independent contribution is needed to describe the state.  
  → This signals **entanglement**.

---

## 🤔 Why It Feels Contradictory
In classical linear algebra, "rank 2" means rows/columns are linearly 
independent, which often suggests separability. But in quantum mechanics, 
the question is different: *can the bipartite state be factorized into 
a product of two subsystems?*  

- Linear independence here corresponds to **multiple Schmidt coefficients**.  
- Multiple Schmidt coefficients = entanglement.  
- Only one Schmidt coefficient = separable.

---

## 📘 Connection to Schmidt Decomposition
The singular values from SVD are the **Schmidt coefficients**:
- One nonzero Schmidt coefficient → separable state.  
- Two nonzero Schmidt coefficients → entangled state.  

So the "weirdness" is just a shift in perspective:  
- **Rank 1** = product state (no entanglement).  
- **Rank > 1** = entangled state.

---

## ✅ Example
- \(|00>): vector `[1,0,0,0]` → reshaped matrix rank = 1 → separable.  
- Bell state \(1/√2(|00> + |11>\): vector `[1/√2,0,0,1/√2]` → reshaped matrix rank = 2 → entangled.
